```python
import papermill as pm
import duckdb, subprocess, os, time
from joblib import Parallel, delayed
```

```python
species_list = [
    "Protonotaria citrea",        # Prothonotary Warbler
    "Archilochus colubris",       # Ruby-throated Hummingbird
    "Meleagris gallopavo",        # Eastern Wild Turkey
    "Aix sponsa",                 # Wood Duck
    "Ursus americanus"            # Black Bear
]
years = range(2017,2018)
year = '2017'
basedir = 'D:/source/statsandmodels/biodiversity'
jobs = 1
model = 'maxent'#['rf', 'xgb','lr']
```

```python
species_list = [s.replace(" ", "_") for s in species_list]
def f(x):
    try:
        spyear = x + '_' + year
        output_path = os.path.join(basedir,f'{model}_{spyear}',f"{model}_{spyear}.tif")
        if os.path.isfile(output_path):
            print(f'{x} exists')
            return (x, 'exists')
        else:
            print('run',x)
            if jobs >1:
                subprocess.run(pm.execute_notebook(f'{model}_model.ipynb','./output/{0}.ipynb'.format(x),parameters=dict(basedir=basedir, sp=x, year=year)), shell=True)
            else:
                pm.execute_notebook(f'{model},'./output/{0}.ipynb'.format(x),parameters=dict(basedir=basedir, sp=x, year=year))
            return (x, 'run complete')
    except Exception as e:
        return (x,'fail', e)
```

```python
completed = Parallel(n_jobs=2, verbose=10)(delayed(f)(species_list[x]) for x in range(len(species_list)))
```
